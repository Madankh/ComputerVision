```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.signal import correlate2d

def normalized_cross_correlation(image, template):
    """
    Compute the normalized cross-correlation to find a template in an image.
    
    Parameters:
    image (numpy.ndarray): The large image to search in (2D array)
    template (numpy.ndarray): The small image region to search for (2D array)
    
    Returns:
    numpy.ndarray: The normalized cross-correlation array
    """
    # Subtract mean from template and image
    template_mean_subtracted = template - np.mean(template)
    image_mean_subtracted = image - np.mean(image)
    
    # Compute the numerator: cross-correlation of mean-subtracted images
    numerator = correlate2d(image_mean_subtracted, template_mean_subtracted, mode='valid')
    
    # Compute the denominator
    image_squared_sum = correlate2d(image_mean_subtracted**2, np.ones(template.shape), mode='valid')
    template_squared_sum = np.sum(template_mean_subtracted**2)
    denominator = np.sqrt(image_squared_sum * template_squared_sum)
    
    # Avoid division by zero
    denominator[denominator == 0] = 1
    
    # Compute normalized cross-correlation
    ncc = numerator / denominator
    
    return ncc

def load_and_preprocess_image(file_path, as_grayscale=True):
    """Load an image and convert it to a numpy array."""
    with Image.open(file_path) as img:
        if as_grayscale:
            img = img.convert('L')
        return np.array(img)

def visualize_results(image, template, ncc):
    """Visualize the original image, template, and NCC result with the best match highlighted."""
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))
    
    # Original image
    ax1.imshow(image, cmap='gray')
    ax1.set_title('Original Image')
    ax1.axis('off')
    
    # Template
    ax2.imshow(template, cmap='gray')
    ax2.set_title('Template to Find')
    ax2.axis('off')
    
    # NCC result
    ax3.imshow(image, cmap='gray')
    ax3.imshow(ncc, cmap='hot', alpha=0.5)
    ax3.set_title('NCC Result Overlay')
    ax3.axis('off')
    
    # Find and highlight the best match
    y, x = np.unravel_index(np.argmax(ncc), ncc.shape)
    rect = plt.Rectangle((x, y), template.shape[1], template.shape[0], 
                         fill=False, edgecolor='lime', linewidth=2)
    ax3.add_patch(rect)
    
    plt.tight_layout()
    plt.show()
    
    return x, y

# Example usage
image_path = 'image.jpeg'  # Replace with your image path
template_path = 'part.png'  # Replace with your template path

# Load and preprocess the image and template
image = load_and_preprocess_image(image_path)
template = load_and_preprocess_image(template_path)

# Compute the normalized cross-correlation
ncc_result = normalized_cross_correlation(image, template)

# Visualize the results and get the best match coordinates
best_x, best_y = visualize_results(image, template, ncc_result)

print(f"Best match found at position: (x={best_x}, y={best_y})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def normalized_cross_correlation(image, template):
    """
    Compute the normalized cross-correlation to find a template in an image.
    
    Parameters:
    image (numpy.ndarray): The large image to search in (2D array)
    template (numpy.ndarray): The small image region to search for (2D array)
    
    Returns:
    numpy.ndarray: The normalized cross-correlation array
    """
    ih, iw = image.shape
    th, tw = template.shape
    
    # Pad the image to handle border conditions
    padded_image = np.pad(image, ((th//2, th//2), (tw//2, tw//2)), mode='constant', constant_values=0)
    
    # Initialize the result array
    ncc = np.zeros((ih, iw))
    
    for i in range(ih):
        for j in range(iw):
            # Extract the image patch
            image_patch = padded_image[i:i+th, j:j+tw]
            
            # Compute the sums for the image patch and template
            sum_image_patch = np.sum(image_patch)
            sum_template = np.sum(template)
            sum_image_patch_sq = np.sum(image_patch**2)
            sum_template_sq = np.sum(template**2)
            sum_product = np.sum(image_patch * template)
            
            # Compute the normalized cross-correlation
            numerator = sum_product
            denominator = np.sqrt(sum_image_patch_sq * sum_template_sq)
            
            if denominator != 0:
                ncc[i, j] = numerator / denominator
    
    return ncc

def load_and_preprocess_image(file_path, as_grayscale=True):
    """Load an image and convert it to a numpy array."""
    with Image.open(file_path) as img:
        if as_grayscale:
            img = img.convert('L')
        return np.array(img)

def visualize_results(image, template, ncc):
    """Visualize the original image, template, and NCC result with the best match highlighted."""
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))
    
    # Original image
    ax1.imshow(image, cmap='gray')
    ax1.set_title('Original Image')
    ax1.axis('off')
    
    # Template
    ax2.imshow(template, cmap='gray')
    ax2.set_title('Template to Find')
    ax2.axis('off')
    
    # NCC result
    ax3.imshow(image, cmap='gray')
    ax3.imshow(ncc, cmap='hot', alpha=0.5)
    ax3.set_title('NCC Result Overlay')
    ax3.axis('off')
    
    # Find and highlight the best match
    y, x = np.unravel_index(np.argmax(ncc), ncc.shape)
    rect = plt.Rectangle((x, y), template.shape[1], template.shape[0], 
                         fill=False, edgecolor='lime', linewidth=2)
    ax3.add_patch(rect)
    
    plt.tight_layout()
    plt.show()
    
    return x, y

# Example usage
image_path = 'image.jpeg'  # Replace with your image path
template_path = 'part.png'  # Replace with your template path

# Load and preprocess the image and template
image = load_and_preprocess_image(image_path)
template = load_and_preprocess_image(template_path)

# Compute the normalized cross-correlation
ncc_result = normalized_cross_correlation(image, template)

# Visualize the results and get the best match coordinates
best_x, best_y = visualize_results(image, template, ncc_result)

print(f"Best match found at position: (x={best_x}, y={best_y})")
```
